```python
# deep learning in perceptron
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_clusters_per_class=1,
                        random_state=41,n_redundant=0,hypercube=False,class_sep=10)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x[:,0],x[:,1],c=y,cmap="winter",s=100)
```

```python
# cteate the function of predict the model
def perceptron(x,y):
  x=np.insert(x,0,1,axis=1)
  weights=np.ones(x.shape[1])
  lr=.1
  for i in range(1000):
     j=np.random.randint(0,100)
     y_hat=step(np.dot(x[j],weights))
     weights=weights + lr*(y[j] - y_hat)*x[1]
  return weights[0],weights[1:]
```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
intercept_,coef_=perceptron(x,y)
```

```python
print(intercept_)
print(coef_)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
plt.figure(figsize=(8,5))
plt.plot(x_input,y_input,color="red",linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap="winter",s=100)
plt.ylim(-3,2)
```
